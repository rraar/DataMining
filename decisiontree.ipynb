```python
#iport library untuk decision tree, dataset, matplotlib
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
#load dataser diabetes
diabetes = datasets.load_diabetes()
features = diabetes['data']
target = diabetes['target']
```

```python
#membuat objek model
decisiontree = DecisionTreeClassifier (random_state=0, max_depth=None,
                                       min_samples_split=2, min_samples_leaf=1,
                                       max_leaf_nodes=None,
                                       min_impurity_decrease=0)
```

```python
#mentraining model decision tree
model = decisiontree.fit(features, target)
```

```python
#mengambil sample observas dan prediksinya
observation = [[9, 8, 7, 6, 5, 4, 3, 2, 1, 0]]
model.predict(observation)
model.predict_proba(observation)
```

```python
#membuat grafik visualisasi decision tree
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import datasets
import matplotlib.pyplot as plt

diabetes = datasets.load_diabetes()
features = diabetes['data']
target = diabetes['target']

decisiontree = DecisionTreeClassifier(random_state=0, max_depth=None,
                                      min_samples_split=2, min_samples_leaf=1,
                                      max_leaf_nodes=None,
                                      min_impurity_decrease=0)

model = decisiontree.fit(features, target)

observation = [[9, 8, 7, 6, 5, 4, 3, 2, 1, 0]]

model.predict(observation)
model.predict_proba(observation)

plt.figure(figsize=(10, 6))
plot_tree(model, feature_names=diabetes.feature_names, filled=True)
plt.savefig('diabetes.png')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
diabetesDataset = pd.read_csv('diabetes.csv')
```

```python
#mengubah kelas (kolom "Species") dari String ke Uniqeue-Integer
diabetesDataset["Outcome"] = pd.factorize(diabetesDataset.Outcome)[0]
```

```python
diabetesDataset.head()
```

```python
#menghapus kolom "id"
irisDataset = diabetesDataset.drop(labels="Id", axis=1)
```

```python
print(irisDataset)
```

```python
#menguabh dataframe ke array Numpy
diabetesDataset = diabetesDataset.to_numpy()
```

```python
print(irisDataset)
```

```python
#membagi dataset => 80 baris data untuk training dan 20 baris data untuk testing
dataTraining = np.concatenate((diabetesDataset[0:400, :], diabetesDataset[500:768, :]),
                              axis=0)
dataTesting = np.concatenate((diabetesDataset[400:500, :], diabetesDataset[500:768, :]),
                              axis=0)
```

```python
#memecah dataset ke input dan label
inputTraining = dataTraining[:, 0:4]
inputTesting = dataTesting[:, 0:4]
labelTraining = dataTraining[:, 4]
labelTesting = dataTesting[:, 4]
print(labelTraining)
len(labelTraining)
```

```python
#mendefinisikan decision tree classifier
model = tree.DecisionTreeClassifier()
```

```python
#mentraining model
model = model.fit(inputTraining, labelTraining)
```

```python
#memprediksi input data testing
hasilPediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi :", hasilPediksi)
```

```python
#menghitung akurasi
prediksiBenar = (hasilPediksi == labelTesting).sum()
prediksiSalah = (hasilPediksi != labelTesting).sum()
print("Prediksi Benar : ", prediksiBenar, "data")
print("Prediksi Salah : ", prediksiSalah, "data")
print("Akurasi : ", prediksiBenar/(prediksiBenar+prediksiSalah)*100, "%")
```
